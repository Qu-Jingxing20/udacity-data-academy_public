```python
train_labels = train.drop(columns = 'comment_text')
train_labels.head(7)
```

```python
### 看2k数据（加快处理时间）
train_labels_2k = train_labels.iloc[:2000, :]
train_labels_2k.shape
```

```python
train_labels_2k.toxic.value_counts()
```

```python
type(train_labels_2k.toxic.value_counts())
```

```python
train_labels_2k.toxic.value_counts()[1]
```

```python
# value counts
## value_counts 适用于 series (一维的计算)
## 可以使用 apply 运用于每列
value_counts = train_labels_2k.apply(pd.value_counts)
value_counts
```

```python
# 计算出ratio
train_labels_2k_ratio = train_labels_2k.apply(pd.value_counts) \
.iloc[-1:,1:]/train_labels_2k.shape[0]
train_labels_2k_ratio
```

```python
# 使用.T转置数据，直接根据ratio画图
train_labels_2k_ratio.iloc[0].sort_values(ascending = False).T \
.plot(kind='bar', legend=False, grid=False);
## iloc[0].sort_values 是按照第0行进行排序
## 因为只按照一行画图，T转置一下
```
