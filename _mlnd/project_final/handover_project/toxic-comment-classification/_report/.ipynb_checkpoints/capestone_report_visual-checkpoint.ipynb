```python
# 用这个框对你计划使用的所有数据包进行设置
# 导入语句
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# classes = ['Toxic','Severe_Toxic','Obscene','Threat','Insult','Identity_Hate']
values = [0.9715,0.9876,0.9860,0.9835,0.9787,0.9752]
```

```python
# s1 = pd.DataFrame(data=pd.Series(values), columns=['Ratio'])
s1 = pd.DataFrame(data=pd.Series(values))
s1.rename(index={0:'Toxic', 1:'Severe Toxic', 2:'Obscene', 3:'Threat', 4:'Insult', 5:'Identity_Hate'}, inplace=True)
s1
```

```python
s1.plot.line(title='Class Ratio', xticks=[0,1,2,3,4,5], legend=False);
```
