> **提示**在你完成数据整理和清理之后，现在可以进行探索性数据分析了。你需要计算统计值，创建可视化图表，解决你在一开始的简介部分中提出的研究问题。我们推荐你采用系统化方法。一次只探索一个变量，然后探索变量之间的关系。

### 研究问题 1（请替换这一标题名称。 ）

```python
# 探索几个数字值列的关系
#数字列共5列有popularity, vote_count, vote_average, budget_adj, revenue_adj
#第一步，先把这5列拿出来做一个新的数据表
list_1 = ['popularity', 'vote_count', 'vote_average', 'budget_adj', 'revenue_adj']
df_tmdb_1 = pd.DataFrame()
for i in list_1:
    df_tmdb_1[i] = df_tmdb[i]
df_tmdb_1.head()
```

```python
#用scatter_matrix显示观察
pd.plotting.scatter_matrix(df_tmdb_1,figsize=(20,20))
```

```python
#除了vote_average 其他四个数据的直方图都非常靠左
df_tmdb_1.describe()
```

```python
#通过describe观察数据，发现budget_adj, revenue_adj 中位数都是0
#再看看每10%的数据分布是多少
def tmdb_quantile (data,n):
    for i in data:
        quantile = data[i].quantile(np.linspace(0,1,n+1))
        print('\n','%',i,'\n',quantile)
tmdb_quantile(df_tmdb_1,10)
```

```python
#找到budget-adj, revenue_adj 任意一个为空值的
#先看看有多少空值
df_tmdb_3 = df_tmdb_1[(df_tmdb_1['budget_adj'] ==0)|(df_tmdb_1['revenue_adj'] ==0)]
df_tmdb_3.shape
```

```python
#除去空值budget-adj, revenue_adj
df_tmdb_2 = df_tmdb_1[(df_tmdb_1['budget_adj'] !=0)&(df_tmdb_1['revenue_adj'] !=0)]
df_tmdb_2.shape
```

```python
tmdb_quantile(df_tmdb_2,10)
```

```python
df_tmdb_2.head(5)
```

```python
#去掉后，调整一下budget_adj, 和 revenue_adj的数据
df_tmdb_copy = df_tmdb_2.copy()
df_tmdb_copy.loc[:,['budget_adj']]=df_tmdb_copy.loc[:,['budget_adj']]/10000000
df_tmdb_copy.loc[:,['revenue_adj']]=df_tmdb_copy.loc[:,['revenue_adj']]/10000000
df_tmdb_copy.head(5)
```

```python
df_tmdb_copy = df_tmdb_copy.rename(columns={'budget_adj': "budget_adj/10^7", 'revenue_adj': 'revenue_adj/10^7'})
df_tmdb_copy.head()
```

```python
pd.plotting.scatter_matrix(df_tmdb_copy,figsize=(20,20))
```

```python
#从df_tmdb_copy的各列关系图看出，vote_count和popularity相关性明显
df_tmdb_copy.plot(x='vote_count', y='popularity',kind='scatter', alpha=0.3)
```

```python
#研究问题1的观察和结论
#1、影片的欢迎程度与影片评论数量正相关，评论数量越多，说明电影手欢迎程度越高；
#2、影片的欢迎程度与影片得分相关性反而并不明显；
#3、结合1、2两个发现，得分高的不一定是受追捧的电影，电影的观看数量反而更能反应电影是否受成功；
#一部评分高的电影不一定受众所有大众，得到各类电影爱好者的喜欢；但是一部评分可以的电影，可能电影本身受众广、引起的话题多，会受到更多的观看；
```

```python
#画图
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
x1 = df_tmdb_copy['vote_count']
y1 = df_tmdb_copy['popularity']
x2 = df_tmdb_copy['vote_average']
y2 = df_tmdb_copy['popularity']
ax1.scatter(x1,y1,color='b',alpha=0.2,label = 'All')
ax2.scatter(x2,y2,color='b',alpha=0.2,label = 'All')
ax1.set_xlabel('vote_count',fontsize=16)
ax1.set_ylabel('Popularity',fontsize=16)
ax2.set_xlabel('vote_average',fontsize=16)
ax2.set_ylabel('Popularity',fontsize=16)
fig.suptitle('Scatter on VoteCount/Pop & VoteAverage/Pop', fontsize=24, color='b')
```

### 研究问题 2（请替换这一标题名称。）

```python
#  研究问题2：投资回报率top5类型电影受欢迎程度和评分情况如何
#首先筛洗要分析的列，并清理数据
list_2 = ['id','genres','popularity', 'vote_average', 'budget_adj', 'revenue_adj']
df_tmdb_genre = pd.DataFrame()
for i in list_2:
    df_tmdb_genre[i] = df_tmdb[i]
df_tmdb_genre.head()
```

```python
#查看缺失值
df_tmdb_genre.isnull().sum()
```

```python
#丢弃genres里的空值并检查
df_tmdb_genre.dropna(axis=0,inplace=True)
df_tmdb_genre.isnull().sum()
```

```python
df_tmdb_genre.info()
```

```python
#去掉budget_adj,revenue_adj为0的值
df_tmdb_genre1 = df_tmdb_genre[(df_tmdb_genre['budget_adj'] !=0)&(df_tmdb_1['revenue_adj'] !=0)]
df_tmdb_genre1.head()
```

```python
#接着处理genres的处理，取每一个数值里第一个genre为每部电影的主要genre，新建一列genre_key，用split处理
df_tmdb_genre1['genre_key'] = df_tmdb_genre1['genres'].apply(lambda x: x.split('|')[0])
df_tmdb_genre1.head()
```

```python
#添加投资回报率ROI列，ROI=benefit/budget_adj*100%
#benefit=revenue_adj-budget_adj
#添加benefit和ROI两列
```

```python
#添加benefit
df_tmdb_genre1['benefit'] = df_tmdb_genre1['revenue_adj']- df_tmdb_genre1['budget_adj']
df_tmdb_genre1.head()
```

```python
#添加ROI列
df_tmdb_genre1['ROI'] = df_tmdb_genre1['benefit']/df_tmdb_genre1['budget_adj']
df_tmdb_genre1.head()
```

```python
df_tmdb_genre1 = df_tmdb_genre1.sort_values(by=['genre_key','ROI'],ascending=[True,False])
df_tmdb_genre1.head(10)
```

```python
#现在需要对genre_key进行groupby处理，得到popularity，vote_average，RIO的mean，并按RIO降序排列，得到top5的数据
df_tmdb_mean = df_tmdb_genre1.groupby('genre_key').mean().sort_values(by=['ROI'],ascending=False)
df_tmdb_mean.head()
```

```python
#发现adventure的ROI特别高，表示怀疑，特意把adcentrue的数据拎出来降序看
df_adv = df_tmdb_genre1.query('genre_key=="Adventure"').sort_values(by=['ROI'],ascending=False)
df_adv.head(10)
```

```python
#观察发现id为8856的ROI数据异常大，将该列删除
df_tmdb_genre2 = df_tmdb_genre1.drop(df_tmdb_genre1.query('id=="8856"').index)
df_tmdb_genre2.shape
```

```python
#检查是否删除
df_tmdb_genre1.shape
```

```python
df_tmdb_genre2.head()
```

```python
#将id，budget_adj,revenue_adj,benefit从数据集丢弃
df_tmdb_genre3 = df_tmdb_genre2.drop(['id','genres','budget_adj','revenue_adj','benefit'], axis=1,inplace=True)
df_tmdb_genre3.head()
```

```python
#再次对genre_key进行groupby处理，得到popularity，vote_average，RIO的mean，并按RIO降序排列，得到top5的数据
df_tmdb_mean2 = df_tmdb_genre3.groupby('genre_key').mean().sort_values(by=['ROI'],ascending=False)
df_tmdb_mean2 = df_tmdb_mean2.rename(columns={'poluparity':'pop_mean','vote_average':'vote_avg_mean','ROI':'ROI_mean'})
df_tmdb_mean.head(10)
```

```python
#找到top5的电影类型，并对列标签重命名
df_tmdb_meantop5 = df_tmdb_mean2.iloc[:5]
df_tmdb_meantop5
```

```python
#画图
sns = set(context='notebook',style='darkgrid')
fig = plt.figure(figsize=(8,8)
ax = fig.add_subplot(1,1,1)
ind = np.arange(len(df_tmdb_meantop5.iloc[0]))    
w = 0.3
ax.bar(ind,df_tmdb_meantop5[0],color='red', alpha=.7, width=w, label="Comedy")
ax.bar(ind,df_tmdb_meantop5[1],color='blue', alpha=.7, width=w, label="Horror") 
ax.bar(ind,df_tmdb_meantop5[2],color='yellow', alpha=.7, width=w, label="Documentary")
ax.bar(ind,df_tmdb_meantop5[3],color='green', alpha=.7, width=w, label="Science Fiction")
ax.bar(ind,df_tmdb_meantop5[4],color='white', alpha=.7, width=w, label="TV Movie")
ax.legend(loc = 'best', fontsize = 12)
ax.set_xticks(ind+4w/2)
mean_label = ['pop_mean','vote_avg_mean','ROI_mean']                
ax.set_xticklabels(mean_lable)
ax.set_ylabel('Values', fontsize=16)
ax.set_xlabel('Features', fontsize=16)
fig.subtitle('Compare Features on Top5 genres', fontsize=24, color='bule')
```
