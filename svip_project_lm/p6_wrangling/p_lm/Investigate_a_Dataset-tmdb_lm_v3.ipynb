### find key genres updated

```python
# 上面的方法有些不严谨
# 可以考虑将分类都拆解出来, 之后进行数量统计
#将众多电影类型分离
df_clean = df1[df1['genres'].notna()]
# 下吗的代码解释
## 删除了 genres 列, 并将后续的结果命名为 genres 并保存 (相当于更新了这列)
## str.split 加了 expand=Ture 参数, 所以会输出很多列
## .stack 将这些列做堆叠合并起来
## reset_index 重置index值, 将 数据堆叠起来(即每个index 对应一个电影的多个genre数据)
df_clean = df_clean.drop('genres', axis=1).join(df_clean['genres'].str.split('|', expand=True).stack().reset_index(level=1, drop=True).rename('genres'))
df_clean['genres'].head(10)
```

```python
# 其实是行数增加了, 接下来对这些行的 genre 做统计就可以了
df1.shape[0], df_clean.shape[0]
```

```python
# 查看统计数量
genres_count = df_clean['genres'].value_counts()
print(genres_count)
type(genres_count), len(genres_count)
```

```python
# 检查图像
## genres_count.index 是说把上面输出的 index 作为 x, 即分类的名字
## genres_count 是把上面输出的值作为 y, 画出高度
### 设定画布大小
plt.figure(figsize=(24,4))
plt.bar(genres_count.index, genres_count);
```

```python
# 现在有了20个分类, 想把各分类对应的 popularity 展现出来
# 因为是20个分类, 所以展现的 popularity 也应该是20个分类的平均值
# 遮掩的话 20个分类 对应 20个值 就可以画出 bar 或者 scatter 图了
# 实现方式是对每一个分类做筛选,之后求平均值, 并将20次平均值存为 y
```
