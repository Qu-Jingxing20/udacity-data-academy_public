## 数据整理


### 常规属性

```python
# 加载数据并打印几行。进行这几项操作，来检查数据
df = pd.read_csv('tmdb-movies.csv')
print(df.columns)
print(df.head(50))
#   类型，以及是否有缺失数据或错误数据的情况。
df.dtypes
df.shape
df.isnull().sum()
round(df.isnull().sum()/df.shape[0],2)
```

从运行结果可以看出，tmdb数据中包含以下数据列：
id,imdb_id,popularity,budget,revenue,original_title,cast,homepage,director,tagline,keywords,overview,runtime,genres,production_companies,release_date,vote_count,vote_average,release_year,budget_adj,revenue_adj.
并且所有21个数据列中，只有cast,homepage,tagline,keywords,production_companies这五列包含有缺失值“NaN”。


### 数据清理



```python
# 在讨论数据结构和需要解决的任何问题之后，
# 在本部分的第二小部分进行这些清理步骤。
# 删除重复值：
print(df.shape)
print(df.duplicated().sum())
df.drop_duplicates(inplace = True)
print(df.shape)
print(df.duplicated().sum())
```

```python
# 清理无用列
# *_adj 考虑了通货膨胀之后的相关电影的预算和收入
# 相比 budget / revenue 更可信，丢弃后者数据
# id 与 imdb_id 作用相同，丢弃后者数据
print(df.shape)
drop_list = ['budget', 'revenue', 'imdb_id']
df.drop(drop_list, axis = 1, inplace = True)
print(df.shape)
df.columns
```

```python
# 调整列
# homepage 缺失较多数值，表示有些电影并没有 homepage
# 有可能增加早期预算，但也有可能会有更高的回报
if 'homepage' in df.columns:
    df['has_homepage'] = df['homepage'].notnull()
    df.drop('homepage', axis = 1, inplace = True)
else:
    print('please pass the data')
print(df.columns[-1:])
df.head()
```

## 探索性数据分析

> **提示**在你完成数据整理和清理之后，现在可以进行探索性数据分析了。你需要计算统计值，创建可视化图表，解决你在一开始的简介部分中提出的研究问题。我们推荐你采用系统化方法。一次只探索一个变量，然后探索变量之间的关系。

### 研究问题 1 - 电影的 popularity 和 vote_count vote_average release_year runtime 之间的关系

```python
# 用这个代码框和其它代码框探索数据。请务必记得添加
#   Markdown 框，以便记录你的观察和调查结果。
# 保留所需数据列
r1list = ['popularity','runtime','vote_count','vote_average','release_year']
# 制作一个新的数据用于scater matirx显示
dfr1 = pd.DataFrame()
for i in r1list:
    dfr1[i] = df[i]
dfr1.head()
```

```python
# 画图观察
# scater matrix图
pd.plotting.scatter_matrix(dfr1, figsize=(20,20));
```

```python
# 看一下图形分布
dfr1.hist(figsize=(10,10));
```

```python
# 发现图形分布比较靠左
dfr1.describe()
```

```python
# min到25%， 25%到50%的数值变化非常小
# runtime有空值，去除空值数据
print(len(dfr1[dfr1['runtime'] == 0]))
print(dfr1.shape)
```

**发现1 vote 和 popularity有关系的:**

```python
# 发现1 vote 和 popularity有关系的
dfr1.plot(x='vote_average',
            y='popularity', kind='scatter', alpha=0.3)
plt.title('Vote vs Popularity');
```

解释：
- 评分和流行程度正相关
- 但也有流行度超过一般水平很多的电影（爆款）
- 评分在比较多在5-8之间，流行度在0-5之间

**发现2 release year 和 popularity有关系的，但它们之间还有其被其它变量影响**

```python
# 发现2 release year 和 popularity有关系的，但它们之间还有其被其它变量影响
dfr1.plot(x='release_year',
            y='popularity', kind='scatter', alpha=0.3)
plt.title('Release vs Popularity');
```

解释：
- 发行年和流行度正相关（发行的时间越晚，流行度越高）
- 原因可能和电影产业的发展壮大和媒体日益增加的宣传效率有关

**发现3 runtime 和 popularity有一些关系，但是没有那么明显**

```python
# 发现3 runtime 和 popularity有一些关系，但是没有那么明显
dfr1.plot(x='runtime',
            y='popularity', kind='scatter', alpha=0.3)
plt.title('Runtime vs Popularity');
```

解释：
- 看电影不是时间越长越好（虽然票价都差不多）
- 从数据看，时常和流行度没有正相关
- 倒是有点像正态分布，在大概90分钟流行度最高
- 在时间很短的那里也有不少数据，可能因为是电影短片的原因

### 研究问题 2 - 电影的 popularity 和 是否有homepage, budget_adj, revenue_adj 的关系

```python
# 去除 budget 空值
r2list = ['popularity','vote_average','budget_adj','revenue_adj', 'has_homepage']
# scater matirx 图示
dfr2 = pd.DataFrame()
for i in r2list:
    dfr2[i] = df[i]
dfr2.head()
```

```python
dfr2.describe()
```

```python
# 根据上面的输出，budget 有很多空值，需要去掉
print(len(dfr2[dfr2['budget_adj'] == 0]))
print(dfr2.shape)

# 以及 revenue 的空值
print(len(dfr2[dfr2['revenue_adj'] == 0]))
print(dfr2.shape)
```

```python
mean = dfr2.groupby('has_homepage').mean()
mean
```

```python
# 柱形图
dfr2.groupby('has_homepage').mean().iloc[0].plot(
    kind = 'bar', alpha = 0.7, color = 'pink')

plt.title('Films with Homepages')
plt.xlabel('Compare Items')
plt.ylabel('Values');
```

```python
# 上图因变量之间数量级差异大，导致变化不明显
mean_edited = mean.copy()

mean_edited['budget_adj'] = mean['budget_adj'] / 10000000
mean_edited['revenue_adj'] = mean['revenue_adj'] / 10000000
mean_edited
```

```python
# 分析在有无 homepage 条件下，popularity 和 revenue 的增长变化幅度
mean_edited.iloc[1] / mean_edited.iloc[0]
```

**研究问题2 发现有自己独立网站的电影在流行度、预算和收入上都明显高于没有独立网站的电影**

```python
# 堆叠柱状图
mean_edited.iloc[0].plot(kind = 'bar', color = 'grey', alpha = 0.4, label = "Homepage No")
mean_edited.iloc[1].plot(kind = 'bar', color = 'pink', alpha = 0.4, label = "Homepage Yes")


plt.legend(loc = 'best', fontsize = 12)

plt.title('Films with / without Homepages')
plt.xlabel('Compare Items')
plt.ylabel('Values');
```

## 结论

### 结论

1. **结论1：**在研究问题1中，我们通过画散点图进行观察 popularity，runtime，vote，release_year 之间的关系，经过分析：电影的得分和流行度正相关；平均投票得分越高，流行度就可能越高；一部电影的运行次数在100左右时，流行度相对更高。

2. **结论2:** 在研究问题2中，我们通过画柱状图进行观察 popularity，budget，revenue，vote 和是否有 homepage 之间的关系，经过分析：一部有主页的电影相比会更受欢迎；制作主页为影片的前期增加不少的投入，但是相对而言，得到的回报更是显著；但是不太影响电影的投票数。

### 分析局限

1. 数据质量。在数据中有大量0预算，0收入的电影。猜测可能是独立电影之类的，这类数据在加到商业电影分析的工作中可能会影响结论。如果后续可以的话，可以考虑将两类电影拆分开或者用新的列区分开。可以让结论更加有针对性。
2. 分析中重点使用的是财务纬度的收入和预算，可以支撑结论。
3. 去掉异常值是为了减少偏差，所以应该根据分析的目的去掉异常值。
